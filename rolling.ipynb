```python
## Import necessary packages
import pandas as pd; 
import numpy as np; 
from tqdm.auto import tqdm;
```

```python
data = pd.read_csv("./data/clean_data.csv"); 
print(data.columns, "\n\n", data.shape)
```

```python
outs = data["Penicillin concentration(P:g/L)"]
data = data[data.columns.drop(list(data.filter(regex='Penicillin')))]
data.head()
```

```python
mean = data.rolling(window=5).mean()
mean = mean.add_suffix("_mean")
std = data.rolling(window=5).std()
std = mean.add_suffix("_std")
maxx = data.rolling(window=5).max()
maxx = mean.add_suffix("_max")
minn = data.rolling(window=5).min()
minn = mean.add_suffix("_min")
```

```python
new_df = pd.concat([mean, std, maxx, minn, pd.Series(outs)], axis=1); 
new_df = new_df.astype('float32')[5:]
new_df.head()
```

```python
new_df.to_csv("./data/ashwin_data.csv", index=False, sep=',')
```
