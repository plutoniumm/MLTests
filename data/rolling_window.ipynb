```python
## Import necessary packages
import pandas as pd; 
data = pd.read_csv("./data/clean_data.csv"); 
print(data.columns, "\n\n", data.shape)
```

```python
outs = data["Penicillin concentration(P:g/L)"]
data = data[data.columns.drop(list(data.filter(regex='Penicillin')))]
data.head()
```

```python
new_df = pd.DataFrame(); 
for attr in ['mean', 'std', 'max', 'min']:
    run = getattr(data.rolling(window=3), attr)(); 
    run = run.add_suffix(f"_{attr}"); 
    print(f"Adding {attr}"); 
    new_df = pd.concat([new_df, run], axis=1); 
```

```python
new_df = pd.concat([new_df, pd.Series(outs)], axis=1); 
new_df = new_df.astype('float32')[5:]
new_df.head()
```

```python
new_df.to_csv("./data/ashwin_data.csv", index=False, sep=',')
```
